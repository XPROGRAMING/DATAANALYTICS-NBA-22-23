```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dados = pd.read_csv('nba.csv')
```

```python
dados.head(10)
```

```python
nomes_das_colunas = dados.columns
nomes_das_colunas
```

#Tratamento de dados

```python
valores_nulos = dados.isnull()
valores_nulos_por_coluna = valores_nulos.sum()
valores_nulos_por_coluna
```

```python
dados = dados.fillna(0)
```

```python
valores_nulos = dados.isnull()
valores_nulos_por_coluna = valores_nulos.sum()
valores_nulos_por_coluna
```

```python
dados.head(10)
```

##FG nescessario para marcar uma sexta de 3p

```python
grupo_por_time = dados.groupby('Team')['3P'].mean()
```

```python
grupo_por_time
```

```python
grupo_por_time = pd.DataFrame(grupo_por_time)
```

```python
grupo_por_time.head(10)
```

```python
grupo_por_time = grupo_por_time.sort_values(by='3P', ascending=False)
top_10_times = grupo_por_time.head(30)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_times.index, y=top_10_times['3P'], palette='rocket')
plt.xlabel('Equipe')
plt.ylabel('FG')
plt.title('Top 30 Média de Tentativas de 3P por Equipe da NBA')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#Média de idade por time



```python
dados.head(10)
```

```python
media_idade_por_time = dados.groupby('Team')['Age'].mean()
media_idade_por_time = media_idade_por_time.round().astype(int)
media_idade_por_time
```

```python
media_idade_por_time = pd.DataFrame(media_idade_por_time)
```

```python
media_idade_por_time
```

```python
media_idade_por_time = media_idade_por_time.sort_values(by='Age', ascending=False)
media_idade = media_idade_por_time.head(30)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x=media_idade.index, y=media_idade['Age'], palette='rocket')
plt.xlabel('Equipe')
plt.ylabel('Idade')
plt.title('Idade média por time')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#Maior pontudor

```python
maior_pontuador = dados.groupby('Player Name')['PTS'].mean()
maior_pontuador
```

```python
maior_pontuador
```

```python
maior_pontuador = maior_pontuador.sort_values(by='PTS', ascending=False)
pontuador = maior_pontuador.head(30)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x=pontuador.index, y=pontuador['PTS'], palette='rocket')
plt.xlabel('Equipe')
plt.ylabel('Jogador')
plt.title('Jogadores com maior PPG')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
